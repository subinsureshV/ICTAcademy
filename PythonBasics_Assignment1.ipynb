**(1).Simple ATM Calculator.**
Write a program that takes the user's account balance and withdrawal amount. If
sufficient balance is available, deduct the amount and display the updated balance.
Otherwise, print an error message. ?

```python
# Step 1: Read account balance from the user
balance = int(input("Enter account balance: "))

# Step 2: Read withdrawal amount from the user
withdrawal = int(input("Enter withdrawal amount: "))

# Step 3: Check if sufficient balance is available
if withdrawal <= balance:

    # Step 4: Deduct withdrawal amount
    balance = balance - withdrawal

    # Step 5: Display successful transaction message
    print("Transaction successful. Remaining balance: ₹", balance)

else:
    # Step 6: Display error message for insufficient balance
    print("Insufficient balance. Transaction failed.")
```

**(2)Electricity Bill Generator.** Create a bill calculator where the user enters the number of units consumed. The rate is
₹5 per unit for the first 100 units, ₹7 for the next 100, and ₹10 for anything above 200
units. Calculate and display the total bill ?

```python
# Step 1: Read the number of units consumed from the user
units = int(input("Enter units consumed: "))

# Step 2: Initialize bill amount
bill = 0

# Step 3: Calculate bill for first 100 units
if units <= 100:
    bill = units * 5

# Step 4: Calculate bill for next 100 units
elif units <= 200:
    bill = (100 * 5) + ((units - 100) * 7)

# Step 5: Calculate bill for units above 200
else:
    bill = (100 * 5) + (100 * 7) + ((units - 200) * 10)

# Step 6: Display the total bill amount
print("Bill amount = ₹", bill)
```

**(3)Temperature Converter.** Write a Python program to convert a given temperature value between Celsius and Fahrenheit. Accept input as a number and a unit indicator ('C' or 'F').?


```python
# Step 1: Read temperature value from the user
temp = float(input("Enter temperature value: "))

# Step 2: Read the unit of temperature (C or F)
unit = input("Enter unit (C/F): ")

# Step 3: Check the unit and perform conversion
if unit == "C":
    # Step 4: Convert Celsius to Fahrenheit
    fahrenheit = (temp * 9/5) + 32
    print("Temperature in Fahrenheit:", fahrenheit, "°F")

elif unit == "F":
    # Step 5: Convert Fahrenheit to Celsius
    celsius = (temp - 32) * 5/9
    print("Temperature in Celsius:", celsius, "°C")

else:
    # Step 6: Handle invalid unit input
    print("Invalid unit. Please enter C or F.")
```

**(4)Mobile Recharge Validity.**
Write a program that takes the recharge date and today’s date as input and checks whether the 28-day recharge plan is still valid or expired?

```python
# Step 1: Import datetime module
from datetime import datetime

# Step 2: Read recharge date from the user
recharge_date = input("Enter recharge date (YYYY-MM-DD): ")

# Step 3: Read today's date from the user
today_date = input("Enter today's date (YYYY-MM-DD): ")

# Step 4: Convert string dates to datetime objects
recharge = datetime.strptime(recharge_date, "%Y-%m-%d")
today = datetime.strptime(today_date, "%Y-%m-%d")

# Step 5: Calculate the number of days passed
days_used = (today - recharge).days

# Step 6: Check if the 28-day plan is valid or expired
if days_used <= 28:
    print("Plan is still valid.")
else:
    print("Plan expired. Recharge again.")
```

**6. Sum of even numbers**
Write a program that takes a number as input and prints the sum of all even numbers from 1 to that number?


```python
# Step 1: Read the number from the user
num = int(input("Enter a number: "))

# Step 2: Initialize sum variable
even_sum = 0

# Step 3: Loop from 1 to the given number
for i in range(1, num + 1):

    # Step 4: Check if the number is even
    if i % 2 == 0:
        even_sum = even_sum + i

# Step 5: Display the result
print("Sum of even numbers:", even_sum)
```

**7. Dictionary – Student Marks Summary**
Write a program that takes input of 3 students’ names and their marks, stores them in a dictionary, and prints the average marks.?

```python
# Step 1: Create an empty dictionary to store student names and marks
students = {}

# Step 2: Read details of 3 students
for i in range(3):

    # Step 3: Read student name
    name = input("Enter name: ")

    # Step 4: Read student marks
    marks = int(input("Enter marks: "))

    # Step 5: Store name and marks in the dictionary
    students[name] = marks

# Step 6: Calculate total marks
total_marks = sum(students.values())

# Step 7: Calculate average marks
average = total_marks / len(students)

# Step 8: Display the average marks
print("Average Marks:", average)
```

**8. Dictionary – Word Frequency**
Write a program that takes a sentence and displays the frequency of each word using a
dictionary

```python
# Step 1: Read a sentence from the user
sentence = input("Enter a sentence: ")

# Step 2: Convert the sentence into a list of words
words = sentence.split()

# Step 3: Create an empty dictionary to store word frequency
frequency = {}

# Step 4: Count frequency of each word
for word in words:
    if word in frequency:
        frequency[word] = frequency[word] + 1
    else:
        frequency[word] = 1

# Step 5: Display the word frequencies
for word, count in frequency.items():
    print(word + ":", count)
```

**9. Functions – Check Prime**
Write a function that checks whether a number is prime. Call the function with user input.


```python
# Step 1: Define a function to check prime number
def is_prime(num):

    # Step 2: Check for numbers less than or equal to 1
    if num <= 1:
        return False

    # Step 3: Check divisibility from 2 to num-1
    for i in range(2, num):
        if num % i == 0:
            return False

    # Step 4: Return True if no divisors found
    return True


# Step 5: Read number from the user
number = int(input("Enter a number : "))

# Step 6: Call the function and display result
if is_prime(number):
    print(number, "is a prime number")
else:
    print(number, "is not a prime number")
```

**10.Vowel Counter**
Write a function that takes a string and returns the number of vowels in it.


```python
# Step 1: Define a function to count vowels
def count_vowels(text):

    # Step 2: Define vowels
    vowels = "aeiouAEIOU"

    # Step 3: Initialize vowel count
    count = 0

    # Step 4: Check each character in the string
    for char in text:
        if char in vowels:
            count = count + 1

    # Step 5: Return the vowel count
    return count


# Step 6: Read string from the user
string = input("Enter a string: ")

# Step 7: Call the function and display result
result = count_vowels(string)
print("Number of vowels:", result)
```

**11.Count Divisible Numbers**
Write a program that takes two numbers as input: a range limit N and a divisor D. Count how many numbers between 1 and N are divisible by D.

```python
# Step 1: Read the range limit from the user
N = int(input("Enter the range limit (N): "))

# Step 2: Read the divisor from the user
D = int(input("Enter the divisor (D): "))

# Step 3: Initialize count variable
count = 0

# Step 4: Loop from 1 to N
for i in range(1, N + 1):

    # Step 5: Check if the number is divisible by D
    if i % D == 0:
        count = count + 1

# Step 6: Display the result
print("Count of numbers divisible by", D, ":", count)
```
