##NumPy (Array Operations)


**Create a NumPy array of shape (5, 5) with random integers between 1 and 50**

```python
import numpy as np

# Create a NumPy array of shape (5, 5) with random integers between 1 and 50
random_array = np.random.randint(1, 51, size=(5, 5))

# Display the array
print(random_array)
```

**Find the row with the maximum sum.**

```python
row_sums = random_array.sum(axis=1)
max_sum_row_index = np.argmax(row_sums)
row_with_max_sum = random_array[max_sum_row_index]

print(f"Row sums: {row_sums}")
print(f"Index of row with maximum sum: {max_sum_row_index}")
print(f"Row with the maximum sum: {row_with_max_sum}")
print(f"Maximum sum: {row_sums[max_sum_row_index]}")
```

**Replace all elements greater than the mean with 0.**

```python
array_mean = np.mean(random_array)
modified_array = random_array.copy() # Create a copy to avoid modifying the original array in place
modified_array[modified_array > array_mean] = 0

print(f"Original array mean: {array_mean:.2f}")
print("Array after replacing elements greater than mean with 0:")
print(modified_array)
```

##Pandas (Data Handling)

**Load the dataset Iris data**

```python
import pandas as pd
import seaborn as sns

# Load the Iris dataset directly using seaborn
iris_df = sns.load_dataset('iris')

print("First 5 rows of the Iris dataset:")
display(iris_df.head())
```

**Display the first 10 rows and basic statistics.**

```python
print("First 10 rows of the Iris dataset:")
display(iris_df.head(10))
```

```python
print("\nBasic statistics of the Iris dataset:")
display(iris_df.describe())
```

**Filter all rows where sepal_length > median and species is "versicolor"**

```python
median_sepal_length = iris_df['sepal_length'].median()

filtered_iris_df = iris_df[
    (iris_df['sepal_length'] > median_sepal_length) &
    (iris_df['species'] == 'versicolor')
]

print(f"Median sepal length: {median_sepal_length:.2f}")
print("\nFiltered DataFrame (sepal_length > median and species == 'versicolor'):")
display(filtered_iris_df)
```

**Group the data by species and compute the mean petal_length.**

```python
mean_petal_length_by_species = iris_df.groupby('species')['petal_length'].mean()

print("Mean petal length by species:")
display(mean_petal_length_by_species)
```

**Add a new column named petal_ratio = petal_length / petal_width.**

```python
iris_df['petal_ratio'] = iris_df['petal_length'] / iris_df['petal_width']

print("DataFrame with new 'petal_ratio' column:")
display(iris_df.head())
```

**Find the top 5 rows with the highest petal_ratio.**

```python
top_5_petal_ratio = iris_df.sort_values(by='petal_ratio', ascending=False).head(5)

print("Top 5 rows with the highest petal_ratio:")
display(top_5_petal_ratio)
```

##Data Visualization


**A histogram of sepal_length with 15 bins**

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.hist(iris_df['sepal_length'], bins=15, edgecolor='black')
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

**A scatter plot of sepal_length vs petal_length, color-coded by species.**

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 7))
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=iris_df, s=100, alpha=0.8)
plt.title('Sepal Length vs Petal Length, Color-coded by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend(title='Species')
plt.grid()
plt.show()
```

**Using Seaborn, plot:**
**A pairplot of the entire dataset.**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a pairplot of the entire dataset, color-coded by species
sns.pairplot(iris_df, hue='species')
plt.suptitle('Pairplot of Iris Dataset', y=1.02) # Add a main title above the subplots
plt.show()
```

**A boxplot showing petal_width distribution for each species.**

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
sns.boxplot(x='species', y='petal_width', data=iris_df)
plt.title('Petal Width Distribution by Species')
plt.xlabel('Species')
plt.ylabel('Petal Width (cm)')
plt.grid(axis='y', alpha=0.75)
plt.show()
```
